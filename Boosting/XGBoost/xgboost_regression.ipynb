```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler, LabelEncoder
from xgboost import XGBRegressor
```

```python
df = pd.read_csv("..\house_prices_practice.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
X = df.drop(['Id', 'SalePrice'], axis=1)
Y = df[['SalePrice']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2, random_state=42)
```

```python
std = StandardScaler()
```

```python
x_train = std.fit_transform(x_train)
x_test = std.transform(x_test)
```

```python
xgb = XGBRegressor()
```

```python
model1 = xgb.fit(x_train, y_train)
```

```python
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
y_pred = model1.predict(x_test)
```

```python
mae1 = mean_absolute_error(y_test, y_pred)
```

```python
param_grid = {
  "n_estimators" : [100, 200, 300, 400],
  "learning_rate" : [.1, .001, 1, .5],
  "max_depth" : [3,5,7,9],
  "subsample" : [.6,.7,.8],
  "colsample_bytree" : [.7,.8,.9],
  "objective" : [ "reg:squarederror"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_CV = GridSearchCV(estimator=xgb, param_grid=param_grid, n_jobs=-1, verbose=10)
```

```python
model2 = grid_CV.fit(x_train, y_train)
```

```python
model2.score(x_train, y_train)
```

```python
model2.score(x_test, y_test)
```

```python
y_pred2 = model2.predict(x_test)
```

```python
mae2 = mean_absolute_error(y_pred2, y_test)
```

```python
print("mae before tuning : ", round(mae1, 2))
print("mae after tuning : ", round(mae2, 2))
```
