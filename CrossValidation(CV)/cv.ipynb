```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
from sklearn.model_selection import train_test_split,KFold, StratifiedKFold, cross_val_score, LeaveOneOut, ShuffleSplit
from sklearn.metrics import accuracy_score
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
x = df.drop(['species'], axis=1)
y = df[['species']]
```

### Hold out cross validation

```python
# Hold out cross validation
x_train, x_test, y_train, y_test = train_test_split(x, y , random_state= 42, test_size= .2)
```

```python
model = dt.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
score1 = accuracy_score(y_test, y_pred)
score1
```

### Stratified K Fold Cross Validation

```python
# k fold cross validation

kfolds = KFold(n_splits=6, shuffle=False)
```

```python
score2 = cross_val_score(dt, x, y, cv = kfolds)
```

```python
score2
```

```python
score2.mean()
```

### Stratified K Fold Cross Validation

```python
strdKfolds = StratifiedKFold(n_splits=6, shuffle=False)
```

```python
score3 = cross_val_score(dt, x, y, cv = strdKfolds)
```

```python
score3
```

```python
score3.mean()
```

### Leave One-Out Cross Validation (LOOCV)

```python
LOO = LeaveOneOut()
```

```python
score4 = cross_val_score(dt, x, y, cv = LOO)
```

```python
len(score4)
```

```python
score4.mean()
```

### Monte Carlo cross-validation

```python
mc_cv = ShuffleSplit(n_splits= 20, test_size=0.2, random_state=42)
```

```python
score5 = cross_val_score(dt, x, y, cv = mc_cv)
```

```python
len(score5)
```

```python
score5.mean()
```
