```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, confusion_matrix
```

```python
df = pd.read_csv("weight-height (1).csv")
df
```

```python
df['Gender'] = df['Gender'].replace({"Male" : 1, "Female" : 0})
```

```python
x = df[['Height','Weight']]
y = df[['Gender']]

```

```python
print(x.shape)
print(y.shape)
```

```python
df.isna().sum()
df.duplicated().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= .2, random_state= 42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(x_test.shape)
```

```python
# creat object and fit the data
los = LogisticRegression()
los_model = los.fit(x_train, y_train)
```

```python
# predict
pred_y = los_model.predict(x_test)
```

```python
# proba_predict
pred_proba = los_model.predict_proba(x_test)
```

```python
pred_y = los_model.predict(x)
pred_proba = los_model.predict_proba(x)
```

```python
df['Female_proba'] = pred_proba[:, 0]
df['Male_proba'] = pred_proba[:, 1]
df['Gender_pred'] = pred_y
```
