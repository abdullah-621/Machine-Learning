```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
df.shape
```

```python
sns.boxplot(df['Age'])
```

```python
Q1 = df['Age'].quantile(.25)
Q3 = df['Age'].quantile(.75)
```

```python
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
```

```python
lower_bound
```

```python
upper_bound
```

```python
df[(df["Age"] <= lower_bound) | (df["Age"] >= upper_bound)].shape
```

```python
df1 = df[(df["Age"] > lower_bound) & (df["Age"] < upper_bound)]
```

```python
df1
```

```python
sns.boxplot(df1['Age'])
```
