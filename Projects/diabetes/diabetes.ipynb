```python
x = df.drop(columns=['Outcome'], axis=1)
y = df['Outcome']
```

```python
MMS = MinMaxScaler()
x = MMS.fit_transform(x)
```

### split the dataset

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state= 42, test_size= .1)
```

### Train and test using LogisticRegression

```python
Log_Re = LogisticRegression()
```

```python
model1 = Log_Re.fit(x_train, y_train)
```

```python
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
pred1 = model1.predict(x_test)
```

```python
print("accuracy_score (LogisticRegression) : ", accuracy_score(pred1, y_test))
```

```python
cm = confusion_matrix(y_test, pred1)
```

```python
proba = model1.predict_proba(x_test)[:,1]
```

```python
accuracy_score(y_pred, y_test)
```
