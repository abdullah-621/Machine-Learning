```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
plt.scatter(df.x, df.y)
```

```python
x = df[['x']]
y = df[['y']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= .5, random_state= 42)
```

```python
x_test.shape
y_test.shape
```

```python
x_train.shape
y_train.shape
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
# plt.figure(figsize=(12,8))
fig, axs = plt.subplots(1, 2, figsize=(12,5))

axs[0].scatter(x_train,y_train, label = "train_data")
axs[0].scatter(x_test, y_test, label = "test data")
axs[0].set_title("Test Data")
axs[0].plot(x_test, y_pred, color = 'red')
axs[0].legend()
# for xi, yi in zip(x_train, y_train):
#     axs[0].text(xi, yi + 0.1, f'({xi},{yi})', ha='center', va='bottom', fontsize=9)



axs[1].scatter(x,y, label = "data set")
axs[1].scatter(x_test, y_pred, label = "test data")
axs[1].set_title("Full data set")
# axs[1].plot(x_test, y_pred, color = 'red')
axs[1].legend()
```
