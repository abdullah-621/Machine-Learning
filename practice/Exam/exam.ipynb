# Model Train

```python
logReg = LogisticRegression()
Knn = KNeighborsClassifier()
tree = DecisionTreeClassifier()
nai_buy1 = GaussianNB()
nai_buy2 = MultinomialNB()
nai_buy3 = BernoulliNB()
nai_buy4 = CategoricalNB()
xgb = XGBClassifier()
adaB = AdaBoostClassifier()
gb = GradientBoostingClassifier()
```

```python
svc = SVC()
```

```python
model_logReg = logReg.fit(x_train, y_train)
model_knn = Knn.fit(x_train, y_train)
model_tree = tree.fit(x_train, y_train)
model_nai_buy1 = nai_buy1.fit(x_train, y_train)
# model_nai_buy2 = nai_buy2.fit(x_train, y_train)
model_nai_buy3 = nai_buy3.fit(x_train, y_train)
# model_nai_buy4 = nai_buy4.fit(x_train, y_train)
model_xgb = xgb.fit(x_train, y_train)
model_adab = adaB.fit(x_train, y_train)
model_gb = gb.fit(x_train, y_train)
```

```python
model_svc = svc.fit(x_train, y_train)
```

```python
print("model_logReg score : ", model_logReg.score(x_train, y_train))
print("model_knn score : ", model_knn.score(x_train, y_train))
print("model_tree score : ", model_tree.score(x_train, y_train))
print("model_nai_buy1 score : ", model_nai_buy1.score(x_train, y_train))
print("model_nai_buy3 score : ", model_nai_buy3.score(x_train, y_train))
print("model_xgb score : ", model_xgb.score(x_train, y_train))
print("model_adab score : ", model_adab.score(x_train, y_train))
print("model_gb score : ", model_gb.score(x_train, y_train))
print("model_svc score : ", model_svc.score(x_train, y_train))
```
