```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_excel("synthetic_health_dataset.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
x = df.drop(['Illness'], axis = 1)
y = df['Illness']
```

```python
# Handle nan value
for col in x.columns:
  if x[col].dtype in ['int64', 'float64']:
    x[col] = x[col].fillna(x[col].mean())
  else:
    x[col] = x[col].fillna(x[col].mode()[0])
```

```python
df.isna().sum()
```

```python
# encoding
x = pd.get_dummies(x)
```

```python
y = LabelEncoder().fit_transform(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state= 42)
```

```python
# scalling
scaler = MinMaxScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# model
Log_reg = LogisticRegression(C=.1)
```

```python
pd.DataFrame(x_train_scaled).isna().sum().sum()
```

```python
model1 = Log_reg.fit(x_train_scaled, y_train)
```

```python
print("Train score : ", model1.score(x_train_scaled, y_train))
```

```python
print("Test score : ", model1.score(x_test_scaled, y_test))
```

```python
pred_y = model1.predict(x_test)
```

```python
print("Accuracy Score : ", accuracy_score(pred_y, y_test))
```

```python
# Use Knn
from sklearn.neighbors import KNeighborsClassifier
Knn = KNeighborsClassifier(n_neighbors=7)
```

```python
model2 = Knn.fit(x_train_scaled, y_train)
```

```python
model2.score(x_train_scaled, y_train)
```

```python
model2.score(x_test_scaled, y_test)
```

```python
pred_y_knn = model2.predict(x_test_scaled)
```

```python
cm = confusion_matrix(pred_y_knn, y_test)
```
