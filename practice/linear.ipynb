```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_absolute_error,r2_score
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
df = df.rename(columns={"x" : "height", "y" : "wieght"})
```

```python
df.head()
```

```python
x = df[['height']]
y = df.drop(['height'], axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y, random_state=42, test_size=.2)
```

```python
mms = MinMaxScaler()
x_train_scaled = mms.fit_transform(x_train)
x_test_scaled = mms.transform(x_test)
```

```python
lr = LinearRegression()
```

```python
model1 = lr.fit(x_train_scaled, y_train)
```

```python
model1.score(x_train_scaled, y_train)
```

```python
model1.score(x_test_scaled, y_test)
```

```python
c = model1.intercept_
m = model1.coef_
```

```python
c, m
```

```python
m * 0.57306547 + c
```

```python
x_test_scaled[0]
```

```python
pred_y = model1.predict(x_test_scaled)
```

```python
pred_y
```

```python
mean_absolute_error(y_test, pred_y)
```
